# API Data retrieval

```python
import requests
import ast
import pandas as pd
import csv
import cleaner
```

```python
# Replace with your actual API key
api_key = open("WGER_KEY.txt", "r").read().strip('\n')
```

```python
def all_exercises_data():
    exercises_list = []
    url = 'https://wger.de/api/v2/exercise/'

    while url:
        response = requests.get(
            url,
            headers={'Authorization': f'Token {api_key}'}
        )
        data = response.json()
        exercises_list.extend(data['results'])
        url = data['next']  # Get the next page URL

    return exercises_list
```

```python
def make_df():
    e_list = all_exercises_data()

    df = pd.DataFrame(e_list)

    return df
```

```python
new_data = make_df()

new_data
```

## API Data column data retrieval

```python
def get_muscle_data():
    url = 'https://wger.de/api/v2/muscle/'
    response = requests.get(url, headers={'Authorization': f'Token {api_key}'})
    data = response.json()
    muscle_data = {item['id']: item['name'] for item in data['results']}
    
    # Handle pagination if there are more pages
    while data['next']:
        response = requests.get(data['next'], headers={'Authorization': f'Token {api_key}'})
        data = response.json()
        muscle_data.update({item['id']: item['name'] for item in data['results']})
        
    return muscle_data
```

```python
def get_equipment_data():
    url = 'https://wger.de/api/v2/equipment/'
    response = requests.get(url, headers={'Authorization': f'Token {api_key}'})
    data = response.json()
    equipment_data = {item['id']: item['name'] for item in data['results']}
    
    # Handle pagination if there are more pages
    while data['next']:
        response = requests.get(data['next'], headers={'Authorization': f'Token {api_key}'})
        data = response.json()
        equipment_data.update({item['id']: item['name'] for item in data['results']})
        
    return equipment_data
```

```python
def get_category_data():
    url = 'https://wger.de/api/v2/exercisecategory/'
    response = requests.get(url, headers={'Authorization': f'Token {api_key}'})
    data = response.json()
    category_data = {item['id']: item['name'] for item in data['results']}
    
    # Handle pagination if there are more pages
    while data['next']:
        response = requests.get(data['next'], headers={'Authorization': f'Token {api_key}'})
        data = response.json()
        category_data.update({item['id']: item['name'] for item in data['results']})
        
    return category_data

```

```python
# Fetch muscle, equipment, and category data
muscle_data = get_muscle_data()
equipment_data = get_equipment_data()
category_data = get_category_data()
```

```python
print("Muscle Data:", muscle_data)
print("Equipment Data:", equipment_data)
print("Category Data:", category_data)
```

```python
df_exercises = pd.read_csv('api_exercises.csv')
```

```python

def replace_ids_with_names(id_list, mapping):
    names = [mapping[int(i)] for i in id_list if int(i) in mapping]
    return names
```

```python
# Function to process the columns
def process_column(column, mapping):
    return column.apply(lambda x: replace_ids_with_names(ast.literal_eval(x), mapping) if pd.notnull(x) else [])
```

```python
# Replace IDs with names in the specified columns
df_exercises['muscles'] = process_column(df_exercises['muscles'], muscle_data)
df_exercises['muscles_secondary'] = process_column(df_exercises['muscles_secondary'], muscle_data)
df_exercises['equipment'] = process_column(df_exercises['equipment'], equipment_data)
```

```python

# Replace category IDs with names
df_exercises['category'] = df_exercises['category'].apply(lambda x: category_data[int(x)] if pd.notnull(x) else None)
```

```python
# Save the updated DataFrame to a new CSV file
df_exercises.to_csv('fixed_api_exercises.csv', index=False)
```

# API Language Translation with Google API

```python
# pip install googletrans==4.0.0-rc1
```

```python
from googletrans import Translator
```

```python
# Initialize the translator
translator = Translator()
```

```python
# Function to translate text
def translate_text(text, src='auto', dest='en'):
    if pd.isna(text):
        return text
    try:
        translated = translator.translate(text, src=src, dest=dest)
        return translated.text
    except Exception as e:
        print(f"Error translating text: {e}")
        return text
```

```python
# Translate exercise names and descriptions
workout_3['name'] = workout_3['name'].apply(translate_text)
```

```python
workout_3['description'] = workout_3['description'].apply(translate_text)
```
